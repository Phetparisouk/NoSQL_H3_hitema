```python
import seaborn as sbn
import pandas as pd
import warnings
import matplotlib
import numpy
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('data/house_pricing.csv')
```

```python
data.head()
```

```python
sbn.relplot(x='LotArea', y='SalePrice', data=data)
```

```python
filter_lot_area_and_price = data[data.LotArea<20000][data.SalePrice<500000]
sbn.relplot(x='LotArea', y='SalePrice', data=filter_lot_area_and_price)
```

```python
sbn.relplot(x='LotFrontage', y='LotArea', data=data)
```

```python
filter_lot_area_and_price = data[data.LotFrontage<200][data.LotArea<100000]
sbn.relplot(x='LotArea', y='SalePrice', data=filter_lot_area_and_price)
```
